```python
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example intent dataset
intent_data = {
    "greeting": ["Hello", "Hi", "Good morning", "Hey there!"],
    "goodbye": ["Goodbye", "See you later", "Bye", "Catch you later"],
    "weather": ["What's the weather?", "Tell me the weather forecast", "Is it sunny today?"]
}

# Prepare sentences and labels
sentences = []
labels = []

for intent, examples in intent_data.items():
    sentences.extend(examples)
    labels.extend([intent] * len(examples))

# Encode the labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Convert sentences to embeddings
sentence_embeddings = model.encode(sentences)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(sentence_embeddings, encoded_labels, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train a Logistic Regression classifier
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

# Test the classifier
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Test accuracy: {accuracy * 100:.2f}%")
```

```python
def predict_intent(sentence):
    # Encode the new sentence
    embedding = model.encode([sentence])

    # Predict the intent
    predicted_label = classifier.predict(embedding)
    predicted_intent = label_encoder.inverse_transform(predicted_label)

    return predicted_intent[0]

# Example test
sentence = "is it winter?"
predicted_intent = predict_intent(sentence)
print(f"Predicted intent: {predicted_intent}")
```
